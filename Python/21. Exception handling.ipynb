```python
# program to print the reciprocal of even numbers

try:
    num = int(input("Enter a number: "))
    assert num % 2 == 0
except Exception as e:
    print(e)
    print("Not an even number!")
else:
    reciprocal = 1/num
    print(reciprocal)
```
